## <font color = 'blue'>reshape用法 </font>
> `reshape(m,n)` 是一種轉置函數，其中m與n代表列與行，把一組資料轉換成行列的行式

```python
A.reshape(100,1)
```

```python
A.reshape(4,25)
```
