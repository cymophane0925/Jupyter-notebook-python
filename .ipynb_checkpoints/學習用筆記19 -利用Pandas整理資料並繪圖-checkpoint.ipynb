```python
df = pd.read_csv("http://bit.ly/uforeports") 
```

```python
df.head()
```

```python
df_state = df.groupby("State").count()
```

```python
df_state
```

```python
df_state.sort_values(by = "Time", ascending=False )
```

```python
df_state #並無更改排序
```

```python
df_state.sort_values(by = "Time", ascending=False,inplace=True ) #inplace 覆蓋的意思
```

```python
df_state #變更排序
```

```python
df_state[:20].Time.plot(kind = "bar") #Kind圖形種類,"bar"長條圖
```
