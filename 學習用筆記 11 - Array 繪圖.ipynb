### <font color = "red">其中"多少點"預設為50。linspace生成出來的東西是array。</font>

```python
x = np.linspace(0, 10, 100) # x = np.linspace(起始值，結束值，多少點)
```

## <font color = "blue">篩選Array裡的元素</font>
>Array  的選項可以利用判斷是來挑選元素 <P>
array 中符合元素條件的array = array[條件]<P>
EX: `L[L<0]`
### <font color = "red" face = 微軟正黑體>Array的選項可以利用判斷是來挑選元素<P>array 中符合元素條件的array = array[條件]</font>


```python
x = np.linspace(-5, 5, 500)
y = np.sinc(x)

plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.plot(x[y>0],y[y>0] , 'o')
```
